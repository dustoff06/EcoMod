## Inheritance

Inheritance allows us to define a class that inherits all the methods and properties from another class.

```python
from datetime import date

class Student:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    @classmethod
    def calculate_age(cls, name, birth_year):
        # calculate age an set it as a age
        # return new object
        return cls(name, date.today().year - birth_year)

    def show(self):
        print(self.name + "'s age is: " + str(self.age))
jessa = Student('Jessa', 20)
jessa.show()

# create new object using the factory method
joy = Student.calculate_age("Joy", 1995)
joy.show()
```

## Polymorphism

Polymorphism allows us to define methods in the child class with the same name as defined in their parent class.

```python
class Bird:
    def intro(self):
        print('There are many types of birds.')
    def flight(self):
        print('Most of the birds can fly but some cannot.')
class Sparrow(Bird):
    def flight(self):
        print('Sparrows can fly.')
class Ostrich(Bird):
    def flight(self):
        print('Ostriches cannot fly.')
obj_bird = Bird()
obj_spr = Sparrow()
obj_ost = Ostrich()
obj_bird.intro()
obj_bird.flight()
obj_spr.intro()
obj_spr.flight()
obj_ost.intro()
obj_ost.flight()
```

# Regression Function

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols


# Add a constant term for the intercept
X = sm.add_constant(cl)

def reg(y, X, data):
    # Create a DataFrame that includes both y and X
    formula = 'y ~ ' + ' + '.join(X.columns)
    # Fit the model
    mod = ols(formula, data=data)
    myfit = mod.fit()
    print(myfit.summary())

    # Perform ANOVA excluding the intercept
    table = sm.stats.anova_lm(myfit, typ=2)  # Type 2 sums of squares exclude the intercept
    print(table)

    return myfit

# Call the function with y and X
X=pd.DataFrame(cl['x'])
y=cl['y']
res=reg(y, X, cl)
```

# Residual Functions

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.stats.stattools import durbin_watson
from statsmodels.stats.diagnostic import het_breuschpagan as hbp
from scipy.stats import shapiro
from statsmodels.compat import lzip


def assumptions(results, X):
      X = sm.add_constant(cl)

      # Normality
      shapiro_test = shapiro(results.resid)
      print(f'Shapiro-Wilk p-value: {shapiro_test.pvalue}')

      # Independence
      dw = durbin_watson(results.resid)
      print(f'Durbin-Watson: {dw}')

      # Homeogeneity
      bp = hbp(results.resid, results.model.exog)
      print(f'Breusch-Pagan p-value: {bp[1]}')


      # Multicollinearity
      if X.shape[1]>2:
        vif_data = pd.DataFrame()
        vif_data["feature"] = X.columns
        vif_data["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
        print(vif_data)

      # Linearity
      plt.scatter(results.fittedvalues, results.resid)
      plt.axhline(y=0, color='r', linestyle='--')
      plt.xlabel('Fitted Values')
      plt.ylabel('Residuals')
      plt.title('Residuals vs Fitted Values')
      plt.show()

      # Normality
      sm.qqplot(results.resid, line='45')
      plt.title('Q-Q Plot')
      plt.show()

      # Potential Outliers
      results.rownames = cl.index
      fig = sm.graphics.influence_plot(results, criterion="cooks")
      fig.tight_layout(pad=1.0)

assumptions(res, X)
```
